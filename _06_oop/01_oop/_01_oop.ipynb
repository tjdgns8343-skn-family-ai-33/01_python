
# OOP
Object-Oriented Programming 객체지향 프로그래밍
객체(메모리조각)를 중심으로 프로그램을 작성하는 방법론

- 변수(속성), 함수(메소드)를 묶어서 하나의 단위로써 프로그래밍
- 클래스를 작성후에 이를 바탕으로 n개의 객체(instance)를 생성할 수 있다.

---
## 객체(Object)
- 인식 가능한 것들(사물, 동물, 개념...)
- 속성(이름, 나이, 주소, 가격, 성분. 원산지...)
- 기능(듣기, 말하기, 운동하기, 게임하기, 숨쉬기...)

### 현실의 객체(Object)
- 객체와 객체간의 상호작용에 의해서 사건이 발생한다.

### OOP
- 현실 객체의 개념을 컴퓨터 속에 그대로 옮겨 놓은 것
- 현실 객체(Object)는 속성/기능으로 이루어짐
    - 속성과 기능을 컴퓨터에 정의
    - 속성과 기능이 정의된 문서 == 클래스(class)

- 클래스를 보고 컴퓨터 메모리에 구현한 것 == 컴퓨터 속 객체(instance)
- instance끼리 호출하면 결과 출력, 반환 등의 사건이 발생함 == OOP

```python
# Class
# == 컴퓨터 속 객체(instace)를 만ㄷ르기 위해 속성과 기능을 정의한 문서
class Person:
    nationality = 'Korean'

    # 생성자 함수
    # - 해당 클래스를 이용해 생성되는 객체의 속성 초기화를 담당
    def __init__(self, name, age):
        # 객체의 속성 초기화
        self.name = name
        self.age = age

    # 객체의 기능 역할 함수 정의
    def greeting(self):
        print(f'Hello. I am {self.name}, {self.age} years old')
```

## 생성자 `__init__`

생성자는 객체가 생성될 때 자동으로 실행되어, 객체가 처음 가져야 할 속성을 초기화하는 메서드이다.

Python에서는 생성자 역할을 하는 메서드를 `__init__`이라는 이름으로 정의한다.
아래 코드에서 `Person('홍길동', 33)`을 실행하면 `__init__(self, name, age)`가 자동으로 호출되고, 전달한 값이 `self.name`, `self.age`에 저장된다.

- `self`: 생성된 인스턴스 자신
- `name`, `age`: 객체를 만들 때 외부에서 전달받는 값
- `self.name = name`: 전달받은 `name` 값을 현재 객체의 속성으로 저장

즉, 생성자는 객체가 만들어지는 순간 필요한 초기 데이터를 세팅하는 초기화 담당 메서드이다.


```python
# 클래스를 이용하여 instance 생성하기
kim = Person('김성훈','28')
# Person 클래스의 생성자를 이용해서 Instance 생성
# kim 변수에는 실제 객체가 아닌 객체를 참조할 수 있는 주소가 저장된다.
print(kim, type(kim))
# <class '__main__.Person'> == 클래스명이 데이터 타입이 된다.
# 클래스 == 사용자 정의 데이터 타입

print('kim.name:', kim.name)
print('kim.age: ', kim.age)
print('kim.nationality: ', kim.nationality)

# 한 번 만든 클래스는 재사용이 가능하다
lee = Person('이순신', 33)
print(lee, type(lee))
print('lee.name:', lee.name)
print('lee.age:', lee.age)
print('lee.nationality:', lee.nationality)
```

## self 이해하기

`self`는 메서드를 호출한 인스턴스 자신을 가리키는 매개변수이다.

아래 두 호출은 같은 의미로 볼 수 있다.

- `hong.greeting()`
- `Person.greeting(hong)`

일반적으로는 첫 번째 방식처럼 인스턴스를 통해 메서드를 호출한다.


```python
print(kim)
kim.greeting()
# kim 변수에 저장된 주소로 찾아가
# 해당 주소에 있는 Person 객체의 greeting()을 호출

print("Person: ", Person)
Person.greeting(kim) # 인수로 Person 타입의 객체 주소가 전달 되어야 한다.
```

## 속성(Field)/메서드(Method)

* 클래스 속성: 인스턴스간에 공유될 수 있는 속성
* 클래스 메서드: 인스턴스간에 공유될 수 있는 메서드
---
* 인스턴스 속성: 인스턴스마다 달라지는 속성(self)
* 인스턴스 메서드: 인스턴스속성에 대해서 처리하는 메서드

---

### 함수(function)과 매서드(method)의 차이점
- 함수: 독립적으로 존재 가능, 함수명() 호출로 기능을 수행하는 코드 블록
- 메서드: 클래스, 또는 객체 내에 정의된 함수
    - 클래스명.메서드명()
    - 참조변수.메서드명() 이런 식으로만 호출 가능
        - 참조변수 == 주소를 저자하는 변수


```python
class SKNStudent:

    # Class 속성: 해당 클래스로 만들어진 모든 인스턴스가 공유하는 값
    course_name = "Family AI Camp"

    # 생성자
    def __init__(self, nth, name):

        # instance 속성: 각 인스턴스(객체)마다 개별적으로 가지는 값
        self.nth = nth
        self.name = name

    # 인스턴스 메서드: 첫 번째 메개변수로 self를 받는 메서드
    # - 해당 인스턴스의 속성을 다룰 수 있음.
    def greeting(self):
        print(f'Hello. {self.nth}th and {self.name}')

    # CLass method: 첫 번째 메개변수로 cls를 받는 메서드
    # - cls == 해당 클래스 == SKNStudent

    @classmethod # Decorator
    def good_morning(cls):
        print(f'{cls.course_name} good morning everyone')

    # 정적 메서드(static method)
    # - self, cls 같은 참조 변수를 받지 않음
    # - 인스턴스 상태, 클래스 상태를 직접 다루지 않는 독립적인 기능에 사용한다.
    @staticmethod # Decorator
    def course_info(campus):
        if campus == '서초':
            print("서초점 주소: 서울시 서초구 ...")
        elif campus == '동작':
            print("동작구 주소: 서울시 동작구 ...")
        else:
            print("잘못 입력")

```

```python
# SKNStudent 객체(인스턴스 생성)
kim = SKNStudent('33','김유진')
choi = SKNStudent('33', '최경돈')
print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

# 인스턴스 속성 변경
kim.nth = '34'
print("--- 인스턴스 속성 변경 후 ---")
print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)

# 클래스 속성 변경
SKNStudent.course_name = "AI Camp"
print("--- After change calss field")
print(kim.nth, kim.name, kim.course_name)
print(choi.nth, choi.name, choi.course_name)
```

```python
# Instance Method check
kim.greeting()
choi.greeting()

# Class method check
kim.good_morning() # 인스턴스.클래스메서드()
choi.good_morning()
SKNStudent.good_morning() # 클래스명.클래스메서드 <-- 권장되는 방식

# Static method check
kim.course_info('서초')
choi.course_info('동작')
SKNStudent.course_info('강남')  # 클래스명.정적메서드명() <-- 권장되는 방식

```

## 클래스 속성과 인스턴스 속성 주의점

클래스 속성은 모든 인스턴스가 공유하는 값이다.
반면 인스턴스에 같은 이름의 속성을 새로 대입하면, 그 인스턴스만의 속성이 생겨 클래스 속성을 가리게 된다.


```python
class Student:
    school_name = "서일초등학교" # 클래스 속성

    def __init__(self, name, grade):
        self.name = name
        self.grade = grade

park = Student("박철수", 3)
kim = Student("김영희",4)

print("park.school_name:", park.school_name)
print("kim.school_name:", kim.school_name)
print("Student.school_name: ", Student.school_name)

# 클래스명.클래스속성 = "값" -> 공유되는ㄴ 클래스 속성 값 수정
Student.school_name = "역삼초등학교"
print("park.school_name:", park.school_name)
print("kim.school_name:", kim.school_name)
print("Student.school_name: ", Student.school_name)

# 인스턴스.클래스속성명 = "값"
# -> 해당 인스턴스에 "클래스속섬명"과 같은 "인스턴스 속성"이 추가된다.
park.school_name = "서이초등학교"
print(park.__dict__)
print(kim.__dict__)
print("park.school_name:", park.school_name)
print("kim.school_name:", kim.school_name)
print("Student.school_name: ", Student.school_name)
```

## 객체지향 원칙
1. 캡슐화
  - 속성/메소드(함수) 묶어서 선언하고
  - 속성의 직접적인 접근을 막고, 메소드를 통한 사용만 허용.
2. 상속
  - 부모클래스에 선언된 속성/메소드를 자식클래스에서 선언없이 사용하는 것.
  - 코드 재사용성이 좋다.
3. 다형성
  - 상속에 기반한 기술. 여러 자식클래스를 부모타입으로 제어하는 것.
4. 추상화
  - 공통된, 보편적인 속성/메소드를 추출해 내는 것. 또는 이를 통해 제어
하는 것.
