# 캡슐화
- 속성/메서드를 하나의 클래스에 함께 선언
- private 속성(변수) 선언해서 외부에서 직접 접근을 막고, 제공된 메소드를 통해서 사용하도록 강제
  - private 변수 선언 방법 : 변수명 앞에 __를 붙인다.
- 캡슐화 원칙을 잘 지켰다면, 주요한 데이터가 외부에 노출되지 않은 은닉화 효과까지 얻게된다.

```python
class Account:
    def __init__(self, username, balance):
        # __가 붙은 속성 == 생성된 인스턴스만 사용 가능한 private 속성
        self.__username = username
        self.__balance = balance

    # 외부에서 private 속성을 다룰 수 있는 메서드
    def info(self): # 인스턴스 메서드
        print(f'{self.__username}님의 현재 잔액은 {self.__balance}원 입니다.')

    def deposit(self, money): # 입금 메서드
        # 오입금 방지(0 이하 입금 막기)
        if money <= 0:
            print("입금액이 유효하지 않습니다")
        else:
            self.__balance += money
            print("입금 금액:", money, 'won')
            print('current balance: ', self.__balance, 'won')

    def withdraw(self, money): # 출금 메서드
        if money <= 0:
            print('출금액이 유효하지 않습니다')
        elif money >= self.__balance:
            print("잔액이 부족합니다")
        else:
            self.__balance -= money
            print('출금 금액:', money, 'won')
            print('current balance: ', self.__balance, 'won')
```

```python
acc1 = Account("홍길동",10_000)
# print(acc1.username)  # private 속성은 외부에서 직접 접근이 불가능하다.
# print(acc1.__balance)
acc1.info()

acc1.deposit(50_000)
acc1.deposit(0)

acc1.withdraw(10_000)
acc1.withdraw(100_000)
acc1.withdraw(0)

acc1.info()
```

## Python private과 name mangling

Python에서 `__balance`처럼 언더스코어 두 개로 시작하는 속성은 외부에서 그대로 접근하기 어렵게 이름이 바뀐다.
이를 name mangling이라고 한다.

다만 이것은 보안을 위한 완전한 접근 제한이라기보다, 클래스 외부에서 직접 사용하지 말라는 강한 신호에 가깝다.


```python
acc = Account('김개똥', 1_000_000)
print(acc.__dict__) # acc가 참조하는 인스턴스에 존재하는 모든 속성을 dict 형태로 반환
```
