# Has-a 포함관계
- 하나의 클래스가 다른 클래스를 속성으로 가지고 있는 경우
- A has a B 성립하는 경우

# Is-a 상속관계
* 하나의 클래스가 다른 클래스를 상속하는 경우
* A is a B 성립하는 경우

## Has-a와 Is-a 선택 기준

상속은 `A is a B`가 자연스럽게 성립할 때 사용한다.
포함관계는 `A has a B`가 자연스럽게 성립할 때 사용한다.

예를 들어 `Taxi is a Car`는 자연스럽지만, `Car is an Engine`은 어색하다.
따라서 자동차가 엔진 기능을 재사용해야 한다면 상속보다 포함관계가 더 적절하다.
