# 운전 콘솔 앱 (OOP)

## 객체지향프로그래밍 OOP
- 현실세계의 모든 사건은 객체(object)와 객체사이의 상호작용인 것에 주목한다.
- 객체간의 상호작용을 구현하기 위해 의인화가 적용된다.
- 객체는 스스로 책임을 가지고 행동을 수행할 수 있다.
- 객체는 적절한 책임을 가질수 있도록 클래스 단위로 잘 분리해 작성해야 한다.
- 단일책임원칙 SRP Single Responsibility Principle 하나의 객체는 하나의 책임만 진다.
- (SOLID 객체지향 개발원칙을 참고)

## 객체간의 상호작용
- 객체는 아는 것(속성)과 할수 있는 것(메서드)으로 구성된다.
- 객체는 자기 할수 있는 것을 메서드로써 외부에 노출한다.
- 객체와 객체는 메세지를 주고받는다.
- 송신자객체는 수신자객체에 매개인자를 통해 메소드를 호출함으로써 메시지를 전송한다.
- 수신자객체는 일련의 책임을 수행후 리턴값으로 송신자객체에 응답한다.

## Driving Application

### 1. 요구사항 명세 (클라이언트)
- 운전프로그램을 만들어 주세요.
- 운전자는 시동걸기/끄기, 악셀 또는 브레이크를 밟을 수 있습니다.
- 자동차는 엔진시작/끝, 가속/감속을 할 수 있습니다.
- 자동차는 처음에 대기상태 있어야 합니다. (시동이 꺼진 상태)
- 자동차는 운전자에 의해 시동이 걸리고, 이미 시동이 걸려있다면, 또 시동을 걸수는 없습니다.
- 운전자가 악셀을 밟으면, 자동차는 10km/h씩 가속할 수 있습니다.
- 최대속도는 200km/h 입니다.
- 운전자가 브레이크를 밟으면, 자동차는 10km/h씩 감속할 수 있습니다.
- 운전자가 시동을 끄면, 자동차는 더이상 움직일 수 없습니다. (시동이 꺼진 상태)
- 자동차가 달리는 동안에는 시동을 끌수 없습니다.

### 3. 구현 포인트

요구사항은 단순히 속도만 올리고 내리는 것이 아니라, 현재 자동차의 상태를 확인하면서 동작을 제한해야 한다.

- 시동이 꺼져 있으면 가속/감속 불가
- 이미 시동이 걸려 있으면 다시 시동 걸기 불가 o
- 속도가 0보다 크면 시동 끄기 불가 o
- 최대 속도를 초과하지 않도록 제한

노트북에서는 `input()`으로 실행을 멈추지 않도록 샘플 메뉴 번호를 전달해 실행한다.


```python
# Car
class Car:

    status = 0
    velocity = 0

    def start_engine(self):
        if self.status == 1:
            print('Already engine running')

        else:
            self.status += 1
            print('Engine started, Current velocity: 0 km/h')

    def stop_engine(self):
        if self.velocity > 0:
            print('You can not turn off the engine, Please break until stop')

        else:
            self.status -= 1
            print('Engine stopped')

    def increase_speed(self):
        if self.status == 0:
            print('Engine is turned off, You cannot accelerate the car')

        elif self.velocity >= 200:
            print('Current velocity is over 200, you cannot accelerate the car')

        else:
            self.velocity += 10
            print(f'Current velocity: {self.velocity} km/h')


    def decrease_speed(self):
        if self.status == 0:
            print('Engine is turned off, You cannot decelerate the car')

        elif self.velocity < 0:
            print('The car is stopped now, you cannot decelerate the car')

        else:
            self.velocity -= 10
            print(f'Current velocity: {self.velocity} km/h')
```

```python
# Driver
class Driver():

    def __init__(self):
        self.__car = Car()

    def start_car(self):
        return self.__car.start_engine()

    def stop_car(self):
        return self.__car.stop_engine()

    def accelerate(self):
        return self.__car.increase_speed()

    def brake(self):
        return self.__car.decrease_speed()

    # 프로그램 종료 시 엔진 꺼짐 확인 해주기 위해 값 가져와 주는 메서드
    def get_value(self):
        return self.__car.status

    # 프로그램 강제 종료 시 값들 초기화 해주는 메서드
    def init_car(self):
        self.__car.status = 0
        self.__car.velocity = 0
```

```python
# UI
class Menu:

    def __init__(self):
        self.__driver = Driver()

    # 프로그램 실행
    def drive_car(self):
        # 시동 켜기
        i = int(input("""시동 켜기 : 1"""))
        while i != 1:
            print('Please start engine first, press number 1')
            i = int(input("""시동 켜기 : 1""")) # 일단 시동부터!!

        self.__driver.start_car()
        # 엔진이 꺼지고 프로그램 종료가능 혹은 강제 종료
        while True:
            i = int(input("""
                시동 켜기 : 1
                시동 끄기 : 2
                가속 : 3
                감속 : 4
                프로그램 종료 : 5
                프로그램 강제종료 : 0"""))

            # 엔진이 안꺼진 상태에서 계속
            if i == 5 and self.__driver.get_value() != 0:
                print('Please end up the program after stop engine')
                continue
            # 엔진 꺼지고 프로그램 종료
            if i == 5 and self.__driver.get_value() == 0:
                break
            # 차량 강제 종료
            if i == 0:
                self.__driver.init_car()  # initialization of car's values
                break

            # 메뉴판
            match i:
                case 1:
                    self.__driver.start_car()
                case 2:
                    self.__driver.stop_car()
                case 3:
                    self.__driver.accelerate()
                case 4:
                    self.__driver.brake()

        # 프로그램 종료 안내
        print('Program ended')
```

```python
menu1 = Menu()
```

```python
menu1.drive_car()
```
