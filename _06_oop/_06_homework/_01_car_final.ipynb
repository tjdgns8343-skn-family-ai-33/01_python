# 운전 콘솔 앱 (OOP)

## 객체지향프로그래밍 OOP
- 현실세계의 모든 사건은 객체(object)와 객체사이의 상호작용인 것에 주목한다.
- 객체간의 상호작용을 구현하기 위해 의인화가 적용된다.
- 객체는 스스로 책임을 가지고 행동을 수행할 수 있다.
- 객체는 적절한 책임을 가질수 있도록 클래스 단위로 잘 분리해 작성해야 한다.
- 단일책임원칙 SRP Single Responsibility Principle 하나의 객체는 하나의 책임만 진다.
- (SOLID 객체지향 개발원칙을 참고)

## 객체간의 상호작용
- 객체는 아는 것(속성)과 할수 있는 것(메서드)으로 구성된다.
- 객체는 자기 할수 있는 것을 메서드로써 외부에 노출한다.
- 객체와 객체는 메세지를 주고받는다.
- 송신자객체는 수신자객체에 매개인자를 통해 메소드를 호출함으로써 메시지를 전송한다.
- 수신자객체는 일련의 책임을 수행후 리턴값으로 송신자객체에 응답한다.

## Driving Application

### 1. 요구사항 명세 (클라이언트)
- 운전프로그램을 만들어 주세요.
- 운전자는 시동걸기/끄기, 악셀 또는 브레이크를 밟을 수 있습니다.
- 자동차는 엔진시작/끝, 가속/감속을 할 수 있습니다.
- 자동차는 처음에 대기상태 있어야 합니다. (시동이 꺼진 상태)
- 자동차는 운전자에 의해 시동이 걸리고, 이미 시동이 걸려있다면, 또 시동을 걸수는 없습니다.
- 운전자가 악셀을 밟으면, 자동차는 10km/h씩 가속할 수 있습니다.
- 최대속도는 200km/h 입니다.
- 운전자가 브레이크를 밟으면, 자동차는 10km/h씩 감속할 수 있습니다.
- 운전자가 시동을 끄면, 자동차는 더이상 움직일 수 없습니다. (시동이 꺼진 상태)
- 자동차가 달리는 동안에는 시동을 끌수 없습니다.

### 2. 객체 도출
- 운전자
- 자동차
- 프로그램메뉴 (사용자 보게될 메뉴, 입력폼, 결과출력 담당할 UI객체)

### 3. 구현 포인트

요구사항은 단순히 속도만 올리고 내리는 것이 아니라, 현재 자동차의 상태를 확인하면서 동작을 제한해야 한다.

- 시동이 꺼져 있으면 가속/감속 불가
- 이미 시동이 걸려 있으면 다시 시동 걸기 불가 o
- 속도가 0보다 크면 시동 끄기 불가 o
- 최대 속도를 초과하지 않도록 제한

노트북에서는 `input()`으로 실행을 멈추지 않도록 샘플 메뉴 번호를 전달해 실행한다.


```python
# Car
class Car:

    def __init__(self):
        self.__status = 0
        self.__velocity = 0

    # 시동 on/off, 속도 확인 하기 위한 메서드
    @property
    def status(self): # get_status()
        return self.__status

    @property
    def velocity(self): # get_status()
        return self.__velocity

    # 자동차 시동 켜기 메서드
    def start_engine(self):
        self.__status += 1
        print('Engine started')

    # 자동차 시동 끄기 메서드
    def stop_engine(self):
        self.__status -= 1
        print('Engine stopped')

    # 자동차 가속 메서드
    def increase_speed(self):
        self.__velocity += 10

    # 운전자 감속 메서드
    def decrease_speed(self):
        self.__velocity -= 10
```

```python
# Driver
class Driver():

    def __init__(self):
        self.__car = Car()

    # 시동 on/off, 속도 확인 하기 위한 메서드
    @property
    def status_car(self):
        return self.__car.status
    @property
    def velocity_car(self):
        return self.__car.velocity

    # 운전자 시동 켜기 메서드
    def start_car(self):
        return self.__car.start_engine()
    # 운전자 시동 끄기 메서드
    def stop_car(self):
        return self.__car.stop_engine()

    # 운전자 가속 메서드
    def accelerate(self):
        return self.__car.increase_speed()

    # 운전자 감속 메서드
    def brake(self):
        return self.__car.decrease_speed()
```

```python
# UI
class Menu:

    def __init__(self):
        self.__driver = Driver()

    # 프로그램 실행 메서드
    def drive_car(self):

        # 시동 키기 전 값들 초기화
        self.__driver.__init__() # 값들 초기화

        # 시동 켜기
        num = int(input("""시동 켜기 : 1"""))
        while num != 1:
            print('Please start engine first, press number 1')
            num = int(input("""시동 켜기 : 1"""))

        self.__driver.start_car() # 일단 시동부터!!

        # 엔진 꺼지고 프로그램 종료 혹은 강제 종료
        while True:

            # 현재 상태 표시
            if self.__driver.status_car == 0:
                print (f'Current velocity: {self.__driver.velocity_car} km/h, Engine status: Off')
            else:
                print (f'Current velocity: {self.__driver.velocity_car} km/h, Engine status: On')

            num = int(input("""
                시동 켜기 : 1
                시동 끄기 : 2
                가속 : 3
                감속 : 4
                프로그램 종료 : 5
                프로그램 강제종료 : 0"""))

            # 엔진이 안꺼진 상태에서 경고문구 안내 후 계속
            if num == 5 and self.__driver.status_car != 0:
                print('Please end up the program after stop engine')
                continue

            # 엔진 꺼지고 프로그램 종료 가능 혹은 강제 종료
            if (num == 5 and self.__driver.status_car == 0) or num == 0:
                break

            # 메뉴판
            match num:
                case 1:
                    if self.__driver.status_car == 1:  # 시동이 켜져 있을 때 시동을 키는 경우
                        print('Already engine running')
                    else:   # 시동 켜짐
                        self.__driver.start_car()
                case 2:
                    if self.__driver.velocity_car > 0:  # 차량이 달리고 있을 때 시동을 끄는 경우
                        print('You cannot turn off the engine, Please break until stop')
                    else:   # 시동 꺼짐
                        self.__driver.stop_car()
                case 3:
                    if self.__driver.velocity_car >= 200:  # 속도가 200이 넘었을 때 가속 하는 경우
                        print('You cannot accelerate over 200')
                    elif self.__driver.status_car == 0: # 시동이 꺼져 있는데 가속하는 경우
                        print('You cannot accelerate while engine stopping')
                    else:   # 악셀
                        self.__driver.accelerate()
                case 4:
                    if self.__driver.status_car == 1 | self.__driver.velocity_car <= 0: # 속도가 0이하인데 감속 하는 경우
                        print('The car is stopped now')
                    elif self.__driver.status_car == 0: # 시동이 꺼져있는데 브레이크를 감속하는 경우
                        print('You cannot decelerate while engine stopping')
                    else:
                        self.__driver.brake()

        # 프로그램 종료 안내
        print('Program ended')
```
